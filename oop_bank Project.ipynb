```python
# Simple Bank Operation using OOP

class Bank:
    bank_name = "Pragati Life Insurance PLC"
    branch = "Kawran Bazar, Dhaka"

    def __init__(self, name, id_no, address):
        self.name = name
        self.id_no = id_no
        self.address = address
        self.balance = 0
        print(f"\nHello {self.name}, your account has been created successfully!\n")

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} BDT deposited successfully!")
        print(f"Current Balance: {self.balance} BDT\n")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"{amount} BDT withdrawn successfully!")
            print(f"Remaining Balance: {self.balance} BDT\n")
        else:
            print("Sorry! Not enough balance.\n")

    def mini_statement(self):
        print("\n====== Mini Statement ======")
        print(f"Account Holder: {self.name}")
        print(f"ID Number: {self.id_no}")
        print(f"Address: {self.address}")
        print(f"Current Balance: {self.balance} BDT")
        print("============================\n")


# Main Program Start
print(f"Welcome to {Bank.bank_name}, {Bank.branch}\n")

name = input("Enter your name: ")
id_no = input("Enter your ID number: ")
address = input("Enter your address: ")

b = Bank(name, id_no, address)

while True:
    print("Select an option:")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Mini Statement")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        amount = float(input("Enter amount to deposit: "))
        b.deposit(amount)

    elif choice == 2:
        amount = float(input("Enter amount to withdraw: "))
        b.withdraw(amount)

    elif choice == 3:
        b.mini_statement()

    elif choice == 4:
        print("\nThank you for using Pragati Life Insurance PLC. Goodbye!\n")
        break

    else:
        print("Invalid option! Please try again.\n")
```
